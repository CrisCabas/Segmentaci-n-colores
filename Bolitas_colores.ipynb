```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from matplotlib import pyplot
from matplotlib import image
from google.colab.patches import cv2_imshow
```

```python
# Punto 1: leer la imagen PELOTASLISAS-COLORES.png y mostrar.
img1 = cv2.imread("PELOTASLISAS-COLORES.png")
plt.axis('off')
# Pasamos a RGB:
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1_rgb)
plt.show
```

```python
# Punto 2:  = identifique solo los contornos de las 6 pelotas presentes en la imagen.
img1_hsv = cv2.cvtColor(img1_rgb, cv2.COLOR_RGB2HSV)
# Hacemos una segmentacion utilizando matlab para obtener los valores limites:
lowery1 = np.array([0,0,255])
uppery1 = np.array([255,30,255])
mask = cv2.inRange(img1_hsv,lowery1,uppery1)
black = cv2.bitwise_and(img1_rgb,img1_rgb, mask = mask)
plt.axis('off')
plt.imshow(black)
plt.show()
```

```python
# realizar detector de bordes con canny edge
edges = cv2.Canny(black, threshold1=500, threshold2=500)
cv2_imshow(edges)
```
